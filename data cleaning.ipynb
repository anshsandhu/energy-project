```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('ny-state-weather-data.csv')

dataset.head()
```

```python
time=dataset.DATE.str.split('T',expand=True)[1]
date=dataset.DATE.str.split('T',expand=True)[0]
dataset['TIME']=time
z=dataset.TIME.str.split(':',expand=True)[0]
dataset['TIME']=z
dataset['DATE']=date
month=dataset.DATE.str.split('-',expand=True)[0]
day=dataset.DATE.str.split('-',expand=True)[1]
dataset['MONTH']=month
dataset['DAY']=day
dataset.__delitem__('DATE')
dataset.head()
```

```python
different_types=[]
for i in dataset['NAME']:
  if i not in different_types:
    different_types.append(i)
print(different_types)
print(len(different_types))
#print(len(dataset['NAME']))
```
