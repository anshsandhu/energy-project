```python
import pandas as pd
import numpy as np
```

```python
df_albany = pd.read_csv('../cleaned-data-frames/albany.csv')
df_buffalo = pd.read_csv('../cleaned-data-frames/buffalo.csv')
df_binghamton = pd.read_csv('../cleaned-data-frames/binghamton.csv')
df_rochester = pd.read_csv('../cleaned-data-frames/rochester.csv')
df_nyc = pd.read_csv('../cleaned-data-frames/nyc.csv')
target = pd.read_csv('../cleaned-data-frames/target-electricity-consumption.csv')
```

```python
temp1 = pd.merge(df_albany,df_buffalo,on=['time','month','day'])
```

```python
temp2 = pd.merge(temp1,df_binghamton, on = ['time','month','day'])
```

```python
temp3 = pd.merge(temp2,df_rochester, on = ['time','month','day'])
```

```python
df_combined = pd.merge(temp3,df_nyc, on = ['time','month','day'])
```
